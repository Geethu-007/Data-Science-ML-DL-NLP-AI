# __Model Training__

```python
df.head()
```

```python
df.drop(['day','month','year'],axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
x = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(x_train.corr(),annot=True)
plt.show()
```

```python
def Correlation(dataset, Threshold):
    col_set = set()
    corr_mat = dataset.corr()
    for i in range(len(corr_mat.columns)):
        for j in range(i):
            if abs(corr_mat.iloc[i,j])>Threshold:
                col_n = corr_mat.columns[i]
                col_set.add(col_n)
    return col_set
```

```python
corr_fea = Correlation(x_train, 0.85)
corr_fea
```

```python
x_train.drop(corr_fea, axis=1, inplace=True)
x_test.drop(corr_fea, axis=1, inplace=True)
x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
x_train_sc.shape, x_test_sc.shape
```

```python
x_train_sc
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title('X_Train before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=x_train_sc)
plt.title('X_Train after Scaling')
plt.show()
```

# __Model Selection__

## ___Linear Regression___

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train_sc, y_train)

y_pred = lr.predict(x_test_sc)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error : ",mse)
print("R2 Score : ",r2)
plt.figure(figsize=(5,3))
plt.scatter(y_test,y_pred)
plt.show()
```

## ___Lasso Regression___

```python
from sklearn.linear_model import Lasso
ls = Lasso()

ls.fit(x_train_sc, y_train)

y_pred = ls.predict(x_test_sc)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error : ",mse)
print("R2 Score : ",r2)
plt.figure(figsize=(5,3))
plt.scatter(y_test,y_pred)
plt.show()
```

## ___Ridge Regression___

```python
from sklearn.linear_model import Ridge
rd = Ridge()

rd.fit(x_train_sc, y_train)

y_pred = rd.predict(x_test_sc)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error : ",mse)
print("R2 Score : ",r2)
plt.figure(figsize=(5,3))
plt.scatter(y_test,y_pred)
plt.show()
```

## ___ElasticNet Regression___

```python
from sklearn.linear_model import ElasticNet
el = ElasticNet()

el.fit(x_train_sc, y_train)

y_pred = el.predict(x_test_sc)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error : ",mse)
print("R2 Score : ",r2)
plt.figure(figsize=(5,3))
plt.scatter(y_test,y_pred)
plt.show()
```
